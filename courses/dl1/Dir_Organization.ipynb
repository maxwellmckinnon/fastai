# Run this script from the dir that the training data lives to randomly subsample a validation set

```python
import os
import glob
import random
import pandas as pd
import shutil

# point to your training images and desired validation folder

# Note, this command is very handy for manualling                  
# shuf -n 10 -e * | xargs -i mv {} path-to-new-folder

data = '/root/notebooks/data_small/plant_seedlings/'
train = 'train'
valid = 'valid'
train_dir = os.path.join(data, train)
val_dir = os.path.join(data, valid)
```

```python
# creates a directory, called 'validation', and into it, creates 10 subdirs, one for each class.
if not os.path.exists(val_dir):
    os.makedirs(val_dir)
```

```python
class_names = [filepath.split('/')[-2:-1][0] for filepath in glob.glob(os.path.join(train_dir, "*", ""))]
class_names
```

```python
[os.mkdir(os.path.join(val_dir, class_name)) for class_name in class_names]
```

```python
glob.glob(os.path.join(os.path.join(train_dir, class_names[0]), "*"))
```

```python
# Move percentage of files to the new directory
PERC_VAL = 20 # between 0 and 100

for class_name in class_names:
    all_files_in_class = glob.glob(os.path.join(os.path.join(train_dir, class_name), "*"))
    num_files = len(all_files_in_class)
    random_sample = random.sample(all_files_in_class, int(num_files*PERC_VAL/100))
    print(f'num_files: {num_files}, len(random_sample): {len(random_sample)}\n')
    for filepath in random_sample:
        shutil.move(filepath, os.path.join(val_dir, class_name, os.path.basename(filepath)))
```
